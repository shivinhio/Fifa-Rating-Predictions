```python
def outlier_treatment(col):
    ''' calculating outlier indices and replacing them with NA  '''
    #Extract quartiles
    q75, q25 = np.percentile(df_test[col], [75 ,25])
    print(q75,q25)
    #Calculate IQR
    iqr = q75 - q25
    #Calculate inner and outer fence
    minimum = q25 - (iqr*1.5)
    maximum = q75 + (iqr*1.5)
    print(minimum,maximum)
    #Replace with NA
    df_test.loc[df_test[col] < minimum,col] = np.nan
    df_test.loc[df_test[col] > maximum,col] = np.nan
```

```python
from sklearn.model_selection import train_test_split
X=df_test.drop(['Overall'],axis=1)
Y=df_test['Overall']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.50,random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree=DecisionTreeRegressor()
dtree.fit(X_train,Y_train)
```

```python
pred_DT=dtree.predict(X_test)
```

```python
def MAPE(Y_test,pred_DT):
    mape=np.mean(np.abs((Y_test-pred_DT)/Y_test)*100)
    return mape
```

```python
MAPE(Y_test,pred_DT)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
def scores(y, y_):
    print('r square  ', metrics.r2_score(y, y_))
    print('Adjusted r square:{}'.format(1 - (1-metrics.r2_score(y, y_))*(len(y)-1)/(len(y)-X_train.shape[1]-1)))
    print('MAPE:{}'.format(np.mean(np.abs((y - y_) / y))*100))
    print('MSE:', metrics.mean_squared_error(y, y_))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y, y_))) 
```

```python
scores(Y_test,pred_DT)
```
